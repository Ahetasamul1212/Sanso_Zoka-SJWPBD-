```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point
from scipy.ndimage import gaussian_filter
import numpy as np
```

```python
file_path = r"E:\SWJP\JICA_Modeler\Datasets\Reach_dta.csv"
data = pd.read_csv(file_path)
```

```python
shapefile_path = r"E:\SWJP\JICA_Modeler\ne_110m_admin_0_countries\ne_110m_admin_0_countries.shp"
world = gpd.read_file(shapefile_path)
```

```python
bangladesh = world[world['NAME'] == 'Bangladesh']
```

```python
geometry = [Point(xy) for xy in zip(data['Longitude'], data['Latitude'])]
geo_df = gpd.GeoDataFrame(data, geometry=geometry)
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title('Fig-1 : Data Points on the Map of Bangladesh')
bangladesh.plot(ax=ax, color='lightblue', edgecolor='black')  # Light blue color for the map
geo_df.plot(ax=ax, markersize=5, color='red', label='Data Points')  # Red dots for data points
```
