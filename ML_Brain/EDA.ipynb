```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("E:\SWJP\ML_Brain\Data_Port\water_potability.csv")
data_cleaned = data.copy()
data_cleaned['ph'].fillna(data['ph'].mean(), inplace=True)
data_cleaned['Sulfate'].fillna(data['Sulfate'].mean(), inplace=True)
data_cleaned['Trihalomethanes'].fillna(data['Trihalomethanes'].mean(), inplace=True)
```

```python
data
```

<h1 style="color:Skyblue;font-weight: Bold;">Bar Plot</h1>

```python
important_features = ['ph', 'Hardness', 'Conductivity', 'Turbidity']
mean_values = data_cleaned.groupby('Potability')[important_features].mean().reset_index()

for feature in important_features:
    plt.figure(figsize=(6, 4))
    sns.barplot(data=mean_values, x='Potability', y=feature, palette='GnBu')
    plt.title(f'Average {feature} by Potability', fontsize=12, weight='bold')
    plt.xlabel('Potability (1 = Potable, 0 = Non-Potable)', fontsize=10)
    plt.ylabel(f'{feature} (Mean)', fontsize=10)
    plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()
```

<h1 style="color:skyblue;font-weight: Bold;">Box Plot</h1>

```python
for feature in important_features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=data_cleaned, x='Potability', y=feature, palette='viridis')
    plt.title(f'{feature} Distribution by Potability', fontsize=12, weight='bold')
    plt.xlabel('Potability (1 = Potable, 0 = Non-Potable)', fontsize=10)
    plt.ylabel(f'{feature}', fontsize=10)
    plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()
```

<h1 style="color:skyblue;font-weight: Bold;">Line Chart Plotting</h1>

```python
for feature in important_features:
    plt.figure(figsize=(6, 4))
    sns.lineplot(data=data_cleaned, x=range(len(data_cleaned)), y=feature, hue='Potability', palette='viridis')
    plt.title(f'{feature} Trends by DO', fontsize=12, weight='bold')
    plt.xlabel('Index', fontsize=10)
    plt.ylabel(f'{feature}', fontsize=10)
    plt.legend(title='Potability', loc='best')
    plt.grid(axis='both', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()
```

<h1 style="color:skyblue;font-weight: Bold">Pie Chart </h1>

```python
potability_counts = data_cleaned['Potability'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(potability_counts, labels=['Non-Potable', 'Potable'], autopct='%1.1f%%', colors=['skyblue', 'orange'], startangle=140)
plt.title('Fig-9 (Proportion of Potable vs Non-Potable Water)', fontsize=12, weight='bold')
plt.tight_layout()
plt.show()
```

<h1 style="color:skyblue;font-weight: Bold;">Scatter Plot</h1>

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=data_cleaned, x='ph', y='Hardness', hue='Potability', palette='viridis', alpha=0.7)
plt.title('Fig-10 (pH vs Hardness by DO)', fontsize=12, weight='bold')
plt.xlabel('pH', fontsize=10)
plt.ylabel('Hardness', fontsize=10)
plt.grid(axis='both', linestyle='--', linewidth=0.5, alpha=0.7)
plt.legend(title='Potability', loc='best')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(data_cleaned[important_features + ['Potability']], hue='Potability', palette='Blues', diag_kind='kde', height=2)
plt.suptitle('Fig-11 Pairplot of Key Features', fontsize=12, weight='bold', y=1.02)
plt.show()
```

```python
important_features = ['ph', 'Hardness', 'Turbidity']
for feature in important_features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=data_cleaned, x=feature, hue='Potability', kde=True, palette='viridis', bins=30, alpha=0.6)
    plt.title(f'Fig-6 {feature} Histogram by DO', fontsize=12, weight='bold')
    plt.xlabel(feature, fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.tight_layout()
    plt.show()
```

<h1 style="color:skyblue;font-weight: Bold;">KDE Plot</h1>

```python
for feature in important_features:
    plt.figure(figsize=(6, 4))
    sns.kdeplot(data=data_cleaned, x=feature, hue='Potability', fill=True, palette='viridis', alpha=0.5)
    plt.title(f'Fig-7 {feature} KDE Plot vs DO', fontsize=12, weight='bold')
    plt.xlabel(feature, fontsize=10)
    plt.ylabel('Density', fontsize=10)
    plt.tight_layout()
    plt.show()
```

<h1 style="color:skyblue;font-weight: Bold;">Swarm PLot</h1>

```python
for feature in ['ph', 'Turbidity']:
    plt.figure(figsize=(6, 4))
    sns.swarmplot(data=data_cleaned, x='Potability', y=feature, palette='viridis', size=3)
    plt.title(f'{feature} Swarm Plot by Potability', fontsize=12, weight='bold')
    plt.xlabel('Potability (1 = Potable, 0 = Non-Potable)', fontsize=10)
    plt.ylabel(feature, fontsize=10)
    plt.tight_layout()
    plt.show()
```
